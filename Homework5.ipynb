```python
# Import NumPy, use one of the NumPy methods and create an array with a shape of (2, 3, 2).
#  You can use the reshape method -- .reshape()

import numpy as np 

array1 = np.array([2, 3, 2])
print(array1)

newShape = array1.reshape(3, -1)
print (newShape)


# Use NumPy .linspace() to create an array with 6 linearly spaced values between 0 and 20

import numpy as np

arr = np.linspace(0,21)
print(arr)

# Make a Deep Copy of the above array
import copy
deep_copy = copy.deepcopy(arr)
print("deep copy:", deep_copy)

# Concatenate these 3 arrays into a new array named 'newArray'...

a1 = ([[25, 16]])
a2 =  ([[11, 2], [13, 4]])
a3 =  ([[7, 81], [5, 6], [11, 12]])

newArray = np.concatenate((a1, a2, a3))
print("Print NewArray...:", newArray)

# Sort 'newArray' in order into 'sortedArray'

sortedArray = np.sort(newArray)
print("Sorted Array...:", sortedArray)


# Combined and sort the following arrays into one called 'comboArray' ...
 
one = ([10, 11, 12, 13, 14, 15, 16, 17])
two = ([20, 21, 22, 23, 24, 25, 26, 27])
three = ([ 0, 1, 2, 3, 4, 5, 6, 7])

comboArray = one+two+three
comboArray.sort()
print("Now it is sorted:", comboArray)


# Take 'comboArray' and perform the following slicing activities:

# print sec1 - the 2nd element

print(comboArray[1])

# print sec2 - all elements from the 3rd element to the last


print("3rd element to the last....", comboArray[2:])

# print sec3 - all elements from the 4th to the 14th elements

```

```python
print(comboArray[3:14])

# print sec4 - the last 6 elements

print(comboArray[-6:])

# print sec5 - all element from #0 up to and including #15, using the negative number method, i.e. taking a section from the end.

print(comboArray[-len(comboArray):-len(comboArray) + 15])


# print sec6 - from #20 every even element to the end

sec6 = comboArray[16: -1:2]
print(sec6)

# print sec7 - from the last element moving forward, every 5th element.

every_5th_element = comboArray[::-5]
print(every_5th_element)

```

```python
#15. Using Series, create a DataFrame that looks like this:

# IngredientsQuantityUnitFlour4cupsMilk1cupEggs2largeSpam1can

import pandas as pd

ingredients = pd.Series(['Flour', 'Milk', 'Eggs', 'Spam'])
quantities = pd.Series([4, 1, 2, 1])
unit = pd.Series(['cups', 'cup', 'large', 'can'])
df = pd.DataFrame({
    'ingredients': ingredients,
    'quantities': quantities,
    'unit': unit
})
print(df)
# Name: Dinner, dtype: object

dinner_series = pd.Series(['Flour', 'Milk', 'Eggs', 'Spam'], name = 'Dinner')

print(dinner_series)
```
